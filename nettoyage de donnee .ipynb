```python
import pandas as pd 
import matplotlib.pyplot as pt 
import seaborn as sns
import numpy as np 
```

```python
df=pd.read_csv('aden02.csv')
```

```python
## le type de donne  
df.dtypes.value_counts()
```

```python

## suppression de  colonne inutile 
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
##  standardiser le format de date 
df['SaleDate']=pd.to_datetime(df['SaleDate'])
```

```python
## on va changer le nom en francaise  pour bien comprendre le  colonne 
df.set_axis(['Id','Idcolis','terres_utliser', 'location_maison','data_vend','prix_vend', 'Reference_legal','vend-vacant','nom_pro' ,'adress_pro' , 'superficie' , 'tax_district'  , 'valeur_terre','valeur_construite' , 'valeur_total' ,'annee_construit' ,'sale_bain' ,' bain_complet' , 'bain_moitie' ], axis=1 , inplace=True)
df.head()
```

```python
## Dans  le colonne vend-vacant  il y a deux categorie yes et no 
df['vend-vacant'].unique()

```

```python
## changer  n  a no et y a yes 
df['vend-vacant'].replace(['N','Y'],['No','Yes'] , inplace=True)
df['vend-vacant'].unique()
```

```python
pd.DataFrame(df['adress_pro'])
```

```python
## diviser le colonne adress_pro
df[['Adress_maison','ville_maison']]=df.location_maison.str.split(',',expand=True)
df.head()
pd.DataFrame(df['ville_maison'])
```

```python
df[['adresspro','villepro','regionpro']]=df.adress_pro.str.split(',',expand=True)
df.head()
pd.DataFrame(df['regionpro'])
```

```python
##  le valeur manquante dans le jeux de donnee 
for i in df.columns :
    pct_missing=np.mean(df[i].isnull())
    print(f'{i} {pct_missing}%')
```
